```python
url = "https://edugo-api.aspiria.ai/aspiria/course-finder/"

headers = {
    "Accept": "application/json, text/plain, */*",
    "Origin": "https://aspiria.ai",
    "Referer": "https://aspiria.ai/",
    "User-Agent": "Mozilla/5.0"
}


params = {
    "page" : 1,
    "page_size" : 16
}

response = requests.get(url,headers=headers, params=params)
data = response.json()


total_pages = data['data']['pagination']['total_pages']
print('total_pages :',total_pages)


for page in range(1,total_pages+1):
    print(f"\nFetching Page {page}...\n")
    params['page'] = page


    response = requests.get(url,headers=headers, params=params)
    data = response.json()


    course = data['data']['university_courses']

    for item in course:
      print(item['title'])
      print("-" * 40)
```
